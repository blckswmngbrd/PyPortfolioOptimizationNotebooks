# Expected Portfolio Return

```python
np.sum(weights * log_returns.mean())*252
```

# Expected Portfolio Variance

```python
np.dot(weights.T,np.dot(log_returns.cov()*252,weights))
```

# Expected Portfolio Volatility

```python
np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252,weights)))
```

```python
port_returns = []
port_volatilities = []

for x in range(1000):
    weights = np.random.random(number_assets)
    weights /= np.sum(weights)
    port_returns.append(np.sum(weights*log_returns.mean())*252)
    port_volatilities.append(np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*252, weights))))
    
    
port_returns, port_volatilities
```

```python
port_returns = np.array(port_returns)
port_volatilities = np.array(port_volatilities)
```

```python
port_returns, port_volatilities
```

```python
portfolios = pd.DataFrame({'Return':port_returns,'Volatility':port_volatilities})
```

```python
portfolios.head()
```

```python
portfolios.tail()
```

```python
portfolios.plot(x='Volatility', y='Return', kind = 'scatter', figsize=(10,6));
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Return')
```
